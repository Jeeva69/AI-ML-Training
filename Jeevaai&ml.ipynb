```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(42)
```

```python
apartments=[f"Apt_{i}" for i in range(1,21)]
```

```python
Dates=pd.date_range(start="2025-07-01",periods=30,freq="D")
```

```python
Data={
    "Date":np.tile(Dates,len(apartments)),
    "Apartment":np.repeat(apartments,len(Dates)),
    "Electricity_usage":np.random.normal(loc=20,scale=5,size=len(Dates)*len(apartments))
}
```

```python
df=pd.DataFrame(Data)
df
```

```python
daily_avg=df.groupby("Date")["Electricity_usage"].mean()
print(daily_avg.head)
```

```python
print(df.dtypes)
```

```python
usage_per_apartment = df.groupby("Apartment")["Electricity_usage"].sum().sort_values(ascending=False)
```

```python
print(usage_per_apartment)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=usage_per_apartment.index,y=usage_per_apartment)
plt.xticks(rotation=90)
plt.title("Total usage per Apartments")
plt.xlabel("Apartments")
plt.ylabel("Total KWH")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=daily_avg.index, y=daily_avg.values)
plt.title("Daily Average Electricity Usage")
plt.xlabel("Date")
plt.ylabel("Average kwh")
plt.tight_layout()
plt.show()
```

```python
# Assuming daily_avg is a DataFrame with index as datetime and a 'value' column
# If the index is datetime, convert to numeric for plotting
x = pd.to_datetime(daily_avg.index)
y = daily_avg.values
z = np.arange(len(daily_avg))  # using index as z-axis

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Convert datetime to float (ordinal format) for x-axis
x_numeric = x.map(pd.Timestamp.toordinal)

# 3D line plot
ax.plot(x_numeric, y, z, label='Electricity Usage', color='blue')

# Format axes
ax.set_title("3D Daily Average Electricity Usage")
ax.set_xlabel("Date")
ax.set_ylabel("Average kWh")
ax.set_zlabel("Index (Time Progression)")

# Optional: Format x-axis back to readable date labels
ax.set_xticks(x_numeric[::len(x)//5])  # reduce ticks to avoid clutter
ax.set_xticklabels(x[::len(x)//5].strftime('%Y-%m-%d'), rotation=30, ha='right')

plt.tight_layout()
plt.show()
```
